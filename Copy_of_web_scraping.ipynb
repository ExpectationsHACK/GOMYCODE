```python
#1.1) Function to get and parse HTML content from a Wikipedia page:
import requests
from bs4 import BeautifulSoup

def get_wikipedia_html(url):
    response = requests.get(url)
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    return soup

#1.2) Function to extract article title:
def get_article_title(soup):
    title = soup.find('title').get_text()
    return title

#1.3) Function to extract article text and headings:
def get_article_text(soup):
    article_text = {}
    article_content = soup.find('div', {'class': 'mw-parser-output'})
    for child in article_content.children:
        if child.name == 'h2':
            section_heading = child.get_text()
            article_text[section_heading] = []
        elif child.name == 'p':
            paragraph_text = child.get_text()
            article_text[section_heading].append(paragraph_text)
    return article_text

#1.4) Function to collect links that redirect to another Wikipedia page:
def get_wikipedia_links(soup):
    links = []
    for link in soup.find_all('a'):
        href = link.get('href')
        if href and href.startswith('/wiki/'):
            links.append('https://en.wikipedia.org/wiki/Python_(programming_language)' + href)
    return links

#1.5) Function to wrap all the previous functions into a single function:
def parse_wikipedia_article(url):
    soup = get_wikipedia_html(url)
    title = get_article_title(soup)
    article_text = get_article_text(soup)
    links = get_wikipedia_links(soup)
    return {'title': title, 'article_text': article_text, 'links': links}

#1.6) Testing the function on a Wikipedia page:
url = 'https://en.wikipedia.org/wiki/Python_(programming_language)'
article_info = parse_wikipedia_article(url)
print(article_info['title'])
print(article_info['article_text'])
print(article_info['links'])
```
